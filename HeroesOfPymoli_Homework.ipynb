## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0,9,14,19,24,29,34,39,100]
age_labels = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
age_df = pd.DataFrame(purchase_data)

# purchase_data_bins = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
# purchase_data_bins.value_counts()

# PurchaseCountsByAgeGroup = purchase_data["Age Group"].value_counts()
# PurchaseCountsByAgeGroupSorted = PurchaseCountsByAgeGroup.sort_index()
# print(PurchaseCountsByAgeGroupSorted)

age_bins_grp = age_df.groupby(pd.cut(age_df["Age"], age_bins, labels=age_labels))

# Create the names for the four bins
age_bins_df = pd.DataFrame({"Purchase Count":age_bins_grp["Price"].count(), 
                            "Average Purchase Price":age_bins_grp["Price"].mean(),
                            "Total Purchase Value":age_bins_grp["Price"].sum(),
                            "Normalized Totals":age_bins_grp["Price"].sum()/age_bins_grp["SN"].nunique()})

age_bins_df["Average Purchase Price"] = age_bins_df["Average Purchase Price"].map("${:.2f}".format)
age_bins_df["Total Purchase Value"] = age_bins_df["Total Purchase Value"].map("${:,.2f}".format)
age_bins_df["Normalized Totals"] = age_bins_df["Normalized Totals"].map("${:,.2f}".format)
age_bins_df




# purchase_data_bins = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
# purchase_data_bins.value_counts()

# PurchaseCountsByAgeGroup = purchase_data["Age Group"].value_counts()
# PurchaseCountsByAgeGroupSorted = PurchaseCountsByAgeGroup.sort_index()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Find most popular items
# Item Id
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value

top_items_df = pd.DataFrame(purchase_data)
top_items_grp = top_items_df.groupby(['Item ID','Item Name'])
top_items_df = pd.DataFrame({"Purchase Count":top_items_grp["Price"].count(), 
                             "Item Price":top_items_grp["Price"].mean(),
                             "Total Purchase Value":top_items_grp["Price"].sum()})


top_items_df = top_items_df.sort_values("Purchase Count", ascending=False)
top_items_df["Item Price"] = top_items_df["Item Price"].map("${:.2f}".format)
top_items_df["Total Purchase Value"] = top_items_df["Total Purchase Value"].map("${:,.2f}".format)

top_items_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
money_items_df = pd.DataFrame(purchase_data)
money_items_grp = money_items_df.groupby(['Item ID','Item Name'])
money_items_df = pd.DataFrame({"Purchase Count":money_items_grp["Price"].count(), 
                             "Item Price":money_items_grp["Price"].mean(),
                             "Total Purchase Value":money_items_grp["Price"].sum()})


money_items_df = money_items_df.sort_values("Total Purchase Value", ascending=False)
money_items_df["Item Price"] = money_items_df["Item Price"].map("${:.2f}".format)
money_items_df["Total Purchase Value"] = money_items_df["Total Purchase Value"].map("${:,.2f}".format)
money_items_df.head()
```
